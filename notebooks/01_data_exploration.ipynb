# ASL Sign Language Interpreter
## 1. Daten-Exploration

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

DATA_DIR = "../data/offline_composited"
```

```python
classes = sorted(os.listdir(DATA_DIR))
print(f"Klassen ({len(classes)}): {classes}\n")

for cls in classes:
    n = len(os.listdir(os.path.join(DATA_DIR, cls)))
    print(f"  {cls}: {n} Bilder")
```

```python
sample_images = os.listdir(os.path.join(DATA_DIR, classes[0]))
img = Image.open(os.path.join(DATA_DIR, classes[0], sample_images[0]))
print(f"Bildgröße: {img.size}")
plt.imshow(img)
plt.title(classes[0])
plt.axis("off")
plt.show()
```

### Beispielbilder pro Klasse

```python
fig, axes = plt.subplots(5, 6, figsize=(15, 12))
axes = axes.flatten()

for i, cls in enumerate(classes[:30]):
    sample = os.listdir(os.path.join(DATA_DIR, cls))[0]
    img = Image.open(os.path.join(DATA_DIR, cls, sample))
    axes[i].imshow(img)
    axes[i].set_title(cls)
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```
